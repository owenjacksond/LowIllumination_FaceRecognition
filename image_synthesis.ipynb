## Image synthesis single image

```python
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt
# Opening Image
im = Image.open("and_exp.5.jpg")
# Creating object of Brightness class
im3 = ImageEnhance.Brightness(im)
im3 = im3.enhance(0.2)
# showing resultant image
plt.imshow(im3)
plt.show()
im3 = im3.save("test/and_20.jpg")
```

## Image synthesis looping

```python
import os
from os import listdir
from os.path import isdir
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt
count = 0
ori_dir = 'grimace/grimace/'
output_dir = 'test/'
for subdir in listdir(ori_dir):
    # path
    path = ori_dir + subdir + '/'
    path2 = listdir(path)
    for i in path2:
        print(i)
        merge_path = str(path + i)
        im = Image.open(merge_path)
        im3 = ImageEnhance.Brightness(im)
        im3 = im3.enhance(0.2)
        if not os.path.exists(output_dir+subdir):
            os.makedirs(output_dir+subdir)
        img = im3.save(output_dir+subdir+'/'+i)
    print("=======end of subdir=======")
```
